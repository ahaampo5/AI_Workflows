#### 1.1 파이썬의 객체지향언어

- class의 용도  
    - 속성과 행동을 가지는 객체를 생성해주는 설계도로써 설명할 수 있다.  
    - 속성은 Attribute라 해석되며 variable로써 저장된다. 그리고 행동은 Action이라 해석되며 method함수로 표현할 수 있다.
    - 중요하게 생각할 것은 데이터를 저장한 채로 저장되며 동시에 데이터 처리를 할 수 있는 것이다.
    _ 또한 큰 장점으로 프로젝트가 커지고 객체마다 겹치는 부분이 많다면 상속시켜 코드를 짧게 작성할 수 있다.


- Visibility  
실제로 코딩을 하면 직접 접근해서 바꾸면 코드에 큰 변화가 일어나기 때문에 직접적으로 바꾸지 못하도록 만드는 경우가 있다.  
attribute에 쓰이며 위의 예시의 경우 self.machine_power의 값은 버튼을 누를 때에만 켜져야 하므로 접근하지 못하도록 만들어야 한다. 따라서 `self.__machine_power`로 고쳐주었으며 값을 구하지 못하는 경우를 알 수 있다.  
하지만 반환값 구하는 경우에 적절하게 사용하면 좋다.


```python
class Machine(object): 

    def __init__(self, wire : bool):
        self.wire = wire
        self.__machine_power = False
        if self.wire == False:
            self.efficiency = 3
        else:
            self.efficiency = 1
            
            
    def connect_wire():
        if self.wire == False:
            self.wire = True
        else:
            print('Already connected')
            
    
    def power_on(self):
        raise NotImplementedError("Subclass must implement abstract method")
        
    
    # Attribute를 사용할 때 조건을 달아서 제한적으로 사용하도록 만드는 decorator가 존재한다.
    @property
    def machine_power(self):
        return self.__machine_power
        
# 유의할 것 super().__init__()함수에 argument 써주어야 하고 상속받은 class는 모든 상속받은 argument까지 모두 써준다.
class Cleaner(Machine):
    
    def __init__(self, wire : bool, name : str, battery : int):
        super().__init__(wire)
        self.name = name
        self.battery = battery
        
            
    def clear(self):
        print('clear')
        self.battery -= self.efficiency
        
    
    def power_on(self):
        print('Button')
        self.__machine_power = True        

cleaner1 = Cleaner(False, 'Ultrapower',35)
cleaner1.clear()
print(cleaner1.efficiency)
print(cleaner1.battery)
cleaner1.power_on()
print(cleaner1.machine_power)
```

#### 1.2 파이썬 함수에 대한 이해와 Decorator

- 파이썬 함수의 특성

  - 파이썬의 모든 함수는 일급함수이다.  
함수 자체가 하나의 객체로 인식되며 반환되거나 파라메터로 사용될 수 있음을 의미한다.  

이처럼 객체를 다루듯 활용될 수 있고 함수 내부에서 활용됨으로써 다양한 활용이 가능하다. 그 중 하나인 Decorator에 대해서 알아보자.

- Decorator  

```python
def star(func):
    def inner(*args, **kwargs):
        print(args[1] * 30)
        func(*args, **kwargs)
        print(args[1] * 30)
    return inner

@star
def printer(msg, mark):
    print(msg)
printer("Hello", "&")
```
