### 1.1 파이썬의 특성

- Variable, Memory  
변수는 메모리 주소를 가지고 있고 변수에 들어가는 값은 메모리 주소에 할당된다.
: 프로그램의 저장소는 SSD지만 프로그램을 실행시키면 RAM으로 저장시키고 CPU가 순차적으로 그 정보를 해석하고 계산한 후에 사용자에게 결과값을 전달해줍니다.  

`특히 파이썬의 경우에는 파이썬이 OS에 변수명 a라는 이름으로 메모리를 잡아달라고 요청을 하고 그 곳에 데이터 값을 저장한다.`  

- Dynamic Typing  
데이터 값을 변수명에 할당하는 순간에 type을 정해줍니다.  
```python
a = 10
b = 'Hello'
```
- -5 부터 256까지의 int 타입 객체는 고정 메모리 주소를 사용한다.  

- Call by Object Reference  

| Call by Value | Call by Reference | Call by Object Reference|
|:---|:---|:---|
| 함수에 값만 넘김 | 함수에 메모리 주소만 넘김 | 객체의 주소가 함수로 전달 |  


````python
def swap_value (x, y):
    temp = x
    x = y
    y = temp

def swap_offset(offset_x, offset_y):
    temp = ex[offset_x]
    ex[offset_x] = ex[offset_y]
    ex[offset_y] = temp
    
def swap_reference (list_ex, offset_x, offset_y): # 기본적으로 코딩구조는 이렇게 가져오는 게 좋다.
    temp = list_ex[offset_x]
    list_ex[offset_x] = list_ex[offset_y]
    list_ex[offset_y] = temp
````

````python
ex = [1,2,3,4,5]
swap_value(ex[0], ex[1])
print(ex)
ex = [1,2,3,4,5]
swap_offset(1, 2)
print(ex)
ex = [1,2,3,4,5]
swap_reference(ex, 3, 4)
print(ex)
````

### 1.2 새로 배운 문법

- iterable한 자료형의 슬라이싱은 범위가 넘어가는 경우 '끝까지','처음부터'의 의미가 된다.

````python
example_list = [1,2,3,4,5,6]
print(example_list[3:10])   # 10이 끝 인덱스를 넘어가지만 끝까지 출력된다.
print(example_list[-10:5])  # -10이 처음 인덱스를 넘어가지만 처음부터 출력된다.
````

- 2차원 List의 카피는 copy 모듈을 사용한다.

````python
import copy
a = [[1,2,3],[4,5,6]]
b = copy.deepcopy(a)
print(b)
````
